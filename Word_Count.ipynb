## Getting the text file from URL

I retrieved the Constitution text file from https://www.constitution.org/cons/constitu.txt using urllib library and saved it as a textfile as Constitution.txt                  

•	Urllib.request – It opens and reads the url.                               
•	Urlretrieve - Retrieves the content present in the url and saves it in a Constitution text file.


```python
#getting the textfile from URl
import urllib.request
urllib.request.urlretrieve("https://www.constitution.org/cons/constitu.txt", "Constitution.txt");
```

### Reading the text file

1) I  declared the variable file to open a file named Constitution.txt. Open takes 2 arguments, the file that       has to open and  operation  to do on the file.                     
2) Here I  used "r" as argument, which indicates the file to read.                  
3) The available option beside "r" are "w" for write and "a" for append.                       
4) Besides the Constitution text file I read the Stop words file which I saved as a textfile in the folder. I took the stop words from the web. Here is the stop words link from https://algs4.cs.princeton.edu/35applications/stopwords.txt

```python
# reading the Constitution text file
file=open("Constitution.txt","r")
constitution=file.read() 
# reading the stop words text file
file2=open("stop_words.txt")
stopwords=file2.read()
```

Importing regular expressions library

The re.sub() function  replaces  any matches in string. punctuation with the second argument, in this case it is empty string.  
String.punctuation contains characters like commas,semicolons and periods.                    
And later on I lower cased all the word present in the text file and saved it in a data variable.

```python
# importing regular expressions and string library
import re
import string
data=fline = re.sub('['+string.punctuation+']', '', constitution).lower()
data
```

### Removing the Stop words using stop words text file

I used list comprehension to write the for loop for removing the stop words. split command splits the words based on space. Using for loop I removed the stop words and saved list of words in a clean_data variable.

List comprehension: It consists of square brackets containing a expression followed by a for clause. It always returns the resulted list.

```python
# removing stop words
data1=data.split()
clean_data=[s for s in data1 if not s in stopwords]
clean_data
```

### Word Count - Counting the number of occurences of a word

From collections library imported the Counter command to count the number of occurences of  each word present in the given list. The output is stored as a dictionary.                                         
A Counter is a dict subclass for counting hashable objects. It is an unordered collection where elements are stored as dictionary keys and their counts are stored as dictionary values.

```python
#counting the words in a given list
from collections import Counter
word_count=Counter(clean_data)
wordcount=dict(word_count)
wordcount
```

## Sorting the  Dictionary values and printing the top 20 words

For loop is used to sort the keys in the dictionary based on values. And keys are sorted in a descending order.

```python
# sorting the words based on values in a dictionary
for key in sorted(wordcount, key=wordcount.get, reverse=True)[:20]:
    #print(key , wordcount[key])
    print(f'{key:<10}   {wordcount[key]: >15d}')
```

## Word Cloud

Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance.

For generating word cloud in Python, I imported matplotlib, pandas and wordcloud modules from python library. After passing the size, background color arguments, then the plt.show command is used to display the image



```python
# imported necessary libaries for the word cloud
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# specifications required to display the image like size,color
wordcloud = WordCloud(background_color="white",max_words= 25,width=1000,height=1000,relative_scaling=1,normalize_plurals=False).generate_from_frequencies(wordcount)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

##   Word Count with stemming

Stemming : It is the process of reducing the infected words into stem,base or root form.
I imported porterstemmer algorithm  from nltk.stem (Natural language tool kit) library and used list comprehension to stem the words.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
 
ps = PorterStemmer()
#lemmatiser = WordNetLemmatizer()
list_stem = [ps.stem(word) for word in clean_data]
list_stem
```

```python
# counting the stem words
from collections import Counter
word_count1=Counter(list_stem)
wordcount1=dict(word_count1)
```

```python
# sorting the words based on key valie pairs and printing top 20 words
for key, value in sorted(wordcount1.items(), key=lambda x: (-x[1], x[0]))[:20]:
      print(key, value)
```

## Lemmatization

The purpose of Lemmatisation is to group together different inflected forms of a word, called lemma. The process is somehow similar to stemming, as it maps several words into one common root. The output of lemmatisation is a proper word, and basic suffix stripping wouldn’t provide the same outcome.

Lemmatisation is closely related to stemming. The difference is that a stemmer operates on a single word without knowledge of the context, and therefore cannot discriminate between words which have different meanings depending on part of speech. However, stemmers are typically easier to implement and run faster, and the reduced accuracy may not matter for some applications.

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
list_lemma = [lemmatizer.lemmatize(word) for word in clean_data]
```
